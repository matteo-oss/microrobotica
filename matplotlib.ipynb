grafico di una funzione:
il computer lo fa calcolando la funzione su tanti punti, non infinito

 Esempio di un grafico a linea: valori che cambiano in base a una x
usati per modelli di machine learning per prevedere un valore futuro, trend


```python
#array di x con n punti
n=100
x=np.linspace(0, 4*np.pi, n)
#campionamento con 10 valori ho pochi valori e fa il campionamento su quei valori e poi li unisce
y1= np.sin(x)
y2=np.cos(x)
```

```python
fig, ax=plt.subplots(figsize=(9,3))
#prende tanti, pararemtri: 
# figsize dimensione figure in pollici (2,54cm), misurs di stampa sul file
# restituisce l'oggetto figura e assi
#creazione grafico:
#tutte le chiamate per modificare il grafico sono fatte su ax
#"r-.", "k--""
color_1=np.random.uniform(0, 1, 3)
color_2=np.random.uniform(0, 1, 3)
ax.plot(x, y1, color=color_1, label="sin(x)", linewidth=6)
ax.plot(x, y2, color=color_2, label="cos(x)", linewidth=6)#la stringa è il colore
#titolo
ax.set_title("Un grafico a linea", fontsize=10)
#etichette
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
ax.grid()
plt.show()#comanda a jupyter di mostrare il grafico
```

Disegnare le funzioni polinomiali:
- $x^2$ 
- $x^3$
- $x^4$ 

```python
punti=np.linspace(-1, 1, 101)
x_2=punti**2
x_3=punti**3
x_4=punti**4

fig, ax=plt.subplots(figsize=(10,20))
ax.plot(punti,x_2, "r", label="$x^2$")
ax.plot(punti,x_3, "g", label="$x^3$")
ax.plot(punti,x_4, "b", label="$x^4$")
ax.set_title("andamento $x^2$, $x^3$, $x^4$")
ax.set_ylabel("y")
ax.set_xlabel("x")
ax.grid()
ax.legend()
plt.savefig("grafico_potenze.pdf", dpi=200)
plt.show()
```

```python
punti=np.linspace(-1, 1, 101)
x_2=punti**2
x_3=punti**3
x_4=punti**4
funzioni=np.array((x_2, x_3, x_4))
labels=["$x^2$", "$x^3$", "$x^4$"]

fig, axs=plt.subplots(nrows=3 ,ncols=1, figsize=(10,20))
for i, funzione in enumerate(funzioni):
    color=np.random.uniform(0,1,3)
    axs[i].plot(punti,funzione, c=color, label=labels[i])
    axs[i].set_title(f"andamento {labels[i]}")
    axs[i].set_ylabel("y")
    axs[i].set_xlabel("x")
    axs[i].grid()
    axs[i].legend()
plt.show()
```

## grafici a dispersione. 
- insieme di punti
- usati per la variazione di una variabile x rispetto ad un'altra y
- utili per capire se una variabile dipende da un'altra: correlazione

```python
np.random.seed(123)
n_partite=20
alice=np.random.randint(1, 7, n_partite)
bob=np.random.randint(1, 7, n_partite)
```

 nello scatter plot è possibile personalizzare ogni punto:
colorare quando vince alice e quando bob

```python

fig, ax=plt.subplots(figsize=(4,4))
ax.grid()
colori= np.array([ "b" for _ in range(n_partite)])
colori[alice>bob]="r"
colori[alice==bob]="y"
ax.scatter(alice, bob, c=colori, s=100, marker="*", zorder=10)#grafico a dispersione: come il punteggio di alice dipende da bob
ax.set_xlabel("punteggio Alice")
ax.set_ylabel("punteggio Bob")

ax.set_xticks([1,2,3,4,5,6])
ax.set_yticks([1,2,3,4,5,6])#disegna le tacche

plt.show()
```
