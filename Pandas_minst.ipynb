Grafico di una figura di 10 righe e 10 colonne.
su ogni riga riportiamo 10 casi di ogni numero

```python
fig, axs=plt.subplots( nrows=10, ncols=10,figsize=(10,10))

for i in range(10):
    cifra=classe_cifra==i
   
       
    cifre_numero=cifre[cifra]
    
    indici=np.random.randint(0,np.size(cifre_numero,axis=0) ,10 )
    
    for j, indice in enumerate(indici):
        axs[i, j].imshow(cifre_numero[indice].reshape((28, -1)),  cmap="binary")
        
plt.show()
    
        
        

    
   
        

        
```
