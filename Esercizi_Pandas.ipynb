Scarica il file neowise json. Esso contiene i dati di asteroidi e comete.
Crea un notebook in cui:
› è creato un dataframe contenente i dati;
› è calcolata la media del periodo orbitale (colonna pe-riod_yr) per ogni categoria di classe orbitale (la colonna orbit_class);
utilizzando i metodi grafici dei dataframe, è tracciato un grafico a dispersione che rappresenti i valori della colonna moid_au rispetto a quelli di period_yr.

```python
neowise=pd.read_json("./neowise.json")
neowise
```

```python
neowise[["period_yr", "orbit_class"]].groupby("orbit_class").mean()
```

```python
neowise[["period_yr", "moid_au"]].info()
neowise
```

```python

neowise_selected=neowise[["period_yr", "moid_au"]].copy()
neowise_selected.dropna(inplace=True)
neowise_selected.info()
```

```python
fig, ax=plt.subplots(figsize=(5,5))
ax.scatter(neowise_selected["moid_au"].values, neowise_selected["period_yr"].values,cmap="magma", c=neowise_selected["moid_au"].values)
ax.set_xlabel("moid_au")
ax.set_ylabel("period_yr")
ax.grid()
ax.set_title("Grafico a Disperiosne moid_au, period_yr")
plt.show()
```

```python
marziani=pd.read_csv("marziani.csv")
marziani.dropna(inplace=True)
```

```python
caratteristiche=["colore", "n_arti", "peso", "altezza", "larghezza"]
n=len(caratteristiche)
fig, axs=plt.subplots(nrows=n, ncols=n, figsize=(20, 20))
colors={"Simmy": np.random.uniform(0,1,3), "Robby": np.random.uniform(0,1,3)}
for i, caratteristica_x in enumerate(caratteristiche):
    for j, caratteristica_y in enumerate(caratteristiche):
            marziani_x_Simmy=marziani[marziani["specie"]=="Simmy"][caratteristica_x].values
            marziani_y_Simmy=marziani[marziani["specie"]=="Simmy"][caratteristica_y].values
            marziani_x_Robbie=marziani[marziani["specie"]=="Robby"][caratteristica_x].values
            marziani_y_Robbie=marziani[marziani["specie"]=="Robby"][caratteristica_y].values
            
        
            if i!=j:


                axs[i, j].scatter(marziani_x_Simmy, marziani_y_Simmy, color=colors["Simmy"], alpha=0.5, label="Simmy")
                axs[i, j].scatter(marziani_x_Simmy, marziani_y_Simmy, color=colors["Robby"], alpha=0.5, label="Robby")
                axs[i, j].set_xlabel(caratteristica_x)
                axs[i, j].set_xlabel(caratteristica_y)
                axs[i, j].legend()
            else:


                axs[i, j].hist(marziani_x_Simmy, color=colors["Simmy"], alpha=0.5)
                
                axs[i, j].hist(marziani_x_Robbie, color=colors["Robby"], alpha=0.5)
                
                axs[i, j].set_xlabel(caratteristica_x)
                axs[i, j].set_ylabel("Frequenza")
            

           
               



            
plt.show()   
    
```
