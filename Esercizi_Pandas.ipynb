### Es 7:
Scarica il file neowise json. Esso contiene i dati di asteroidi e comete.
Crea un notebook in cui:
› è creato un dataframe contenente i dati;
› è calcolata la media del periodo orbitale (colonna pe-riod_yr) per ogni categoria di classe orbitale (la colonna orbit_class);
utilizzando i metodi grafici dei dataframe, è tracciato un grafico a dispersione che rappresenti i valori della colonna moid_au rispetto a quelli di period_yr.

```python
neowise=pd.read_json("./neowise.json")
neowise
```

```python
neowise[["period_yr", "orbit_class"]].groupby("orbit_class").mean()
```

```python
neowise[["period_yr", "moid_au"]].info()
neowise
```

```python

neowise_selected=neowise[["period_yr", "moid_au"]].copy()
neowise_selected.dropna(inplace=True)
neowise_selected.info()
```

```python
fig, ax=plt.subplots(figsize=(5,5))
ax.scatter(neowise_selected["moid_au"].values, neowise_selected["period_yr"].values)
ax.set_xlabel("moid_au")
ax.set_ylabel("period_yr")
plt.show()
```
