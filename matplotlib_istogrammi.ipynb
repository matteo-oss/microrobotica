```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
dati=pd.read_csv("./Fuel_Consumption_2000-2022.csv")
#tipo dataframe: contiene la tabella
lista_campi=list(dati.columns)

```

```python
for riga in lista_campi:
    fig, ax=plt.subplots(figsize=(15,5))
    valori=dati[str(riga)].values
    
    valori_unici=np.unique(valori)
    ax.hist(valori, bins=valori_unici.size)
    ax.set_xticklabels(valori_unici,rotation=70)
    ax.set_title(riga)


    ax.axis("off")


    
    plt.show()



    
```

### Istogramma dei consumi:
istogramma non è il diagramma a barre:
l'istogramma ha sulla x il consumo, suddivisio in intervalli e per ogni intervallo conta quante macchine hanno il consumo in quell'intervallo

```python
fig, ax=plt.subplots(figsize=(6,2))
consumi=dati["FUEL CONSUMPTION"].values
ax.hist(consumi, bins=50)#bins è il numero di barre, è possibile mettere la lista degli intervalli
ax.set_xlabel("consumi [$l/100km$]")
ax.set_ylabel("N° auto")
ax.grid()
plt.show()
```

## Grafico di una matrice

```python
n=10
matrice=np.random.uniform(0, 1,(n, n))
matrice[0]=0
matrice[-1]=1
```

```python
fig, ax=plt.subplots(figsize=(3,3))
ax.imshow(matrice, cmap="hsv")#mappa ciclica (hsv): i colori stanno su un cerchio: il colore più intenso è anche quello meno intensp
plt.show()
```

matplotlib cerca il minimo e il massimo e li associa i valori
è possibile cambiare la color map.

Una matrice è un'immagine

```python
immagine= plt.imread("./nuovo_sfondo.jpg")
immagine.shape #numero colonne, numero righe, numero matrici: 3->rgb, 4->rgba: a di alpha 
fig, ax=plt.subplots(figsize=(5,5))
ax.imshow(immagine[:,:,0], cmap="Reds_r")#rosso invertito, prendo tutte le righe e tute le colonne con la 1 colonna dei colori:r
plt.show()
```

```python
componente_red=immagine[:,:,0].copy()
```

```python
np.min(componente_red)
```

```python
componente_red[componente_red>255*0.5]=255
fig, ax=plt.subplots(figsize=(5,5))
ax.imshow(componente_red, cmap="Reds_r")#rosso invertito, prendo tutte le righe e tute le colonne con la 1 colonna dei colori:r
plt.show()
```

## Più grafici in una figura

```python
fig, axs=plt.subplots(nrows=1, ncols=2, figsize=(6,3))
print(axs)#lista di assi: su ognuno di essi posso assegnare un'immagine
axs[0].imshow(immagine)
axs[1].imshow(componente_red, cmap="Reds_r")
plt.show()
```

1 figura con 2 sottografici

```python
fig, axs=plt.subplots(nrows=2, ncols=2, figsize=(10,3))
componente_green=immagine[:,:,1].copy()
componente_blue=immagine[:,:,2].copy()

axs[0,0].imshow(immagine)
axs[0, 1].imshow(componente_red, cmap="Reds_r")
axs[1, 0].imshow(componente_green, cmap="Greens_r")
axs[1, 1].hist(componente_blue.flatten())
plt.show()
```
