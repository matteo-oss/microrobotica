Utilizza matplotlib creare un grafico a dispersione che mostri la correlazione tra IMDb Score e Awards Received. I punti del grafico devono avere un colore che dipenda dal numero di voti ricevuti IMDb Votes. I grafici devono avere titolo, griglia ed etichette sugli assi.

```python
data["IMDb Votes"]
```

```python
x=data["IMDb Score"].values
y=data["Awards Received"].values
c=data["IMDb Votes"].values
fig, ax=plt.subplots(figsize=(6,6))
ax.set_xlabel("IMDb Score")
ax.set_ylabel("Awards Received")
ax.grid()
ax.set_title("grafico a dispersione che mostri la correlazione tra IMDb Score e Awards Received")
ax.scatter(x, y, c=c, cmap="hsv", marker="*")
plt.show()
```

Traccia lo stesso grafico facendo in modo che la dimensione dei punti sia proporzionale al numero di voti IMDb Votes.

```python
x=data["IMDb Score"].values
y=data["Awards Received"].values
size=data["IMDb Votes"].values
fig, ax=plt.subplots(figsize=(5,5))
ax.set_xlabel("IMDb Score")
ax.set_ylabel("Awards Received")
ax.grid()
ax.set_title("grafico a dispersione che mostri la correlazione tra IMDb Score e Awards Received")
ax.scatter(x, y, c=c, s=size/2000, cmap="hsv", marker="*")
plt.show()
```

6) Traccia in un'unica figura due due istogrammi che rappresentino rispettivamente le distribuzioni di `IMDb Score` e `Awards Received`.

```python
nrows=2
fig, axs=plt.subplots(ncols=1, nrows=nrows, figsize=(15,15))
valori=[data["IMDb Score"].values, data["Awards Received"].values]
labels=["IMDb Score", "Awards Received"]
for i in range(nrows):
    axs[i].hist(valori[i], color=np.random.uniform(0,1,3), label=labels[i], bins=30)
    axs[i].legend()
    axs[i].set_title(f"istogramma di: {labels[i]}")
    axs[i].set_xlabel(labels[i])
    axs[i].set_ylabel(f"frequenza di:{labels[i]}")
    axs[i].grid()
axs[0].set_xlim([0, 10])
axs[1].set_xlim([0, 100])

plt.show()
    
```
