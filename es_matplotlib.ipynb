### Es 6:

Cerca su Internet degli open data sulla serie storica areer-le della popolazione mondiale degli ultimi 50 anni. Sorica i dati in un file csv e crea un grafico avente come ascissa l'anno e come ordinata il valore della popolazione.

```python
anni=[]
dati=[]
file=open("./population.csv", "r")
righe=file.readlines()
file.close()
for riga in righe:
    campi=riga.split(",")
    if campi[0].lower()=="world":
        anni.append(int(campi[2]))
        dati.append(int(campi[3]))
anni=np.array(anni)
dati=np.array(dati)
```

```python

fig, ax=plt.subplots(figsize=(10,10))

ax.plot(anni, dati, "r",label="popolazione" , linewidth=2)
ax.set_title("Andamento popolazione mondiale 1973-2023",fontsize=10)
ax.set_xlabel("anni")
ax.set_ylabel("popolazione")
ax.legend()
plt.show()
```

### Es 12:

Utilizza numpy per creare una matrice 100 x 100 di numeri interi compresi tra 0 e 255. Crea una figura con due grafici, cioè due sottografici:
› il primo rappresenta la matrice con la mappa di colore che preferisci;
il secondo rappresenta la distribuzione dei valori degli elementi della matrice come un istogramma.

```python
dim=100
matrice=np.random.uniform(100, 10,(dim, dim))
```

```python
fig, axs=plt.subplots(nrows=1, ncols=2,figsize=(20, 10))
axs[0].imshow(matrice, cmap="plasma")#mappa ciclica (hsv): i colori stanno su un cerchio: il colore più intenso è anche quello meno intensp
axs[1].hist(matrice.flatten(), bins=np.size(np.unique(matrice.flatten())))#bins è il numero di barre, è possibile mettere la lista degli intervalli
axs[1].set_xlabel("intensità pixel")
axs[0].axis("off")
axs[1].set_ylabel("frequenza intensità ")
axs[1].grid()
plt.show()
```

### es 10:

Cerca in internet degli open data sulle popolazioni di tutti i comuni italiani. Scarica un file csv contenente le popolazioni di tutti i comuni italiani e:
› inserisci i dati all'interno di un opportuno dizionario di liste;
rappresenta la distribuzione dei valori di popolazione
mediante un istogramma.

```python
dati=pd.read_csv("./popolazione_rese_dal_2018_al_2021 (1).csv")
popolazione=dati["pop_res_18"].values
comuni=dati["comune"].values
```

```python
dizionario_comuni={"comune":[]}
abitanti={"numero_abitanti":[]}

for comune, cittadini in zip(comuni, popolazione):
    dizionario_comuni["comune"].append(str(comune))
    abitanti["numero_abitanti"].append(float(cittadini))


    
        
```

```python
lista_abitanti=[]
for chiave, elemento in abitanti.items():
    
    
    lista_abitanti.append(elemento)
    
array_abitanti=np.array(lista_abitanti)
array_abitanti=array_abitanti.flatten()
array_abitanti=array_abitanti[array_abitanti!=np.isnan(array_abitanti)]
array_abitanti
```

```python
fig, ax=plt.subplots(figsize=(6,6))
color=np.random.uniform(0, 1, 3)
ax.hist(array_abitanti, bins=2000, color=color)
ax.set_xlim([0, 20000])
ax.set_ylabel("Numero comuni")
ax.set_xlabel("popolazione")
plt.show()
```
