```python
penguin = pd.read_csv("./penguins_size.csv")#data è un oggetto DataFrame
penguin
```

```python
penguin["species"]
```

```python
fig, ax=plt.subplots(figsize=(6,6))

colors=[]
dizionario_specie={}
for  specie in penguin["species"]:
    if specie not in dizionario_specie:
        colore=np.array([np.random.uniform(0,1), np.random.uniform(0,1),np.random.uniform(0,1)])
        dizionario_specie[specie]=colore
        colors.append(colore)
    else:
        colors.append(dizionario_specie[specie])
        

array_colors=np.array(colors)

for i, specie in enumerate(penguin["species"].unique()):
    color=array_colors[array_colors==dizionario_specie[specie]]
    mat_color=color.reshape(-1,3)
    penguin_selected_x=penguin["culmen_depth_mm"][penguin["species"]==specie]
    penguin_selected_y=penguin["culmen_length_mm"][penguin["species"]==specie]
    ax.scatter(penguin_selected_x, penguin_selected_y, label=str(specie), c=mat_color)
   
  
ax.set_title("Pinguini")
ax.legend(loc=3)
plt.show()
```

```python
fig, axs=plt.subplots(nrows=3, ncols=1, figsize=(10,10))
colors=[]
dizionario_specie={}
for  specie in penguin["species"]:
    if specie not in dizionario_specie:
        colore=np.array([np.random.uniform(0,1), np.random.uniform(0,1),np.random.uniform(0,1)])
        dizionario_specie[specie]=colore
        colors.append(colore)
    else:
        colors.append(dizionario_specie[specie])
        

array_colors=np.array(colors)

for i, specie in enumerate(penguin["species"].unique()):
    color=array_colors[array_colors==dizionario_specie[specie]]
    mat_color=color.reshape(-1,3)
    penguin_selected_x=penguin["culmen_depth_mm"][penguin["species"]==specie]
    penguin_selected_y=penguin["culmen_length_mm"][penguin["species"]==specie]
    axs[i].scatter(penguin_selected_x, penguin_selected_y, label=str(specie), c=mat_color, alpha=0.5)
    
   
  
    axs[i].set_title(f"{specie}")
    axs[i].legend(loc=4)
plt.show()
```
