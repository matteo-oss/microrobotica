```python
data = pd.read_csv("./penguins_size.csv")#data è un oggetto DataFrame
data
```

```python
data_selected=data[["species", "body_mass_g"]].copy()
data_selected.dropna(inplace=True)
data_selected
```

```python
data_selected.info()
```

```python
data_selected.to_csv("pinguini_selected.csv")
```

### Filtri:
- condizioni:
nome_df[(nome_df["nome_colonna"]condizione)]


- aggiungere una colonna:
data_selected["nome_nuova_colonna"] = data_selected["nome_colonna"]:calcolo

```python
data_selected[(data_selected["body_mass_g"]<=3000) &
             (data_selected["species"]=="Chinstrap")]
```

```python
data_selected["body_mass_kg"] = data_selected["body_mass_g"]/1000
data_selected
```

### Esercizio:
figura con 3 grafici, ogni grafico è un istogramma con le 3 specie con il peso.
Specie:
- Adelle
- Gentoo
- Chinstrap

```python
fig, axs= plt.subplots(figsize=(10,3),ncols=3, nrows=1)
for i, specie in enumerate(data_selected["species"].unique()):
    data_specie=data_selected[data_selected["species"]==specie]
    axs[i].set_title(f"Specie: {specie}")
    axs[i].set_xlabel("Peso pinguini")
    axs[i].set_ylabel("Numero pinguini")

    axs[i].hist(data_specie["body_mass_kg"].values)
plt.show()
    
    
```
