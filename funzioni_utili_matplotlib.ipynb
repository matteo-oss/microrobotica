###  Introduzione:

funzione:

- plot ():
rappresenta un array di ordinate y in funzione di un array di ascisse x, tramite linee oppure tramite punti

- Parametri:
    - marker: stile dei punti
    - color: colore linea
    - alpha: trsparenza
    - linestyle: stile linea
    - linewidth: spesso linea
    - label: etichetta, nome che diamo alla funzione
    - y
    - x
- scatter ():
rappresenta un array di ordinate y in funzione di un array di ascisse x tramite punti in un diagramma a dispersione con la possibilità di personalizzare lo stile di ciascun punto

- Parametri:
    - alpha: trasperenza
    - marker:  stile dei punti
    - label: etichetta
    - х

creare un foglio con i grafici delle funzioni polinomiali dal secondo al quarto grado:

    punti=np.linspace(-1, 1, 101)
    x_2=punti**2
    x_3=punti**3
    x_4=punti**4
    funzioni=np.array((x_2, x_3, x_4))
    labels=["$x^2$", "$x^3$", "$x^4$"]
    colors=["r", "g", "y"]
    fig, axs=plt.subplots(nrows=3 ,ncols=1, figsize=(10,20))
    for i, funzione in enumerate(funzioni):
        axs[i].plot(punti,funzione, c=colors[i], label=labels[i])
        axs[i].set_title(f"andamento {labels[i]}")
        axs[i].set_ylabel("y")
        axs[i].set_xlabel("x")
        axs[i].grid()
        axs[i].legend()
    plt.show()
